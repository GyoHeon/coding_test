```python
bracket = input()
length = len(bracket)
```

```python
if length % 2 == 1: 
  print(0)
  exit(0)
else:
  bracket = bracket.replace('()', '2')
  bracket = bracket.replace('[]', '3')
  bracket = list(bracket)
```

```python
def score(n, array):
  sec = array[n]
  thi = array[n+1]
  if sec.isdigit():
    if thi.isdigit():
      array[n : n+2] = [str(int(sec) + int(thi))]
      return array
    elif not thi.isalnum():
      if array[n-1] == '(' and thi == ')':
        array[n-1:n+2] = [str(int(sec) * 2)]
        return array
      elif array[n-1] == '[' and thi == ']':
        array[n-1:n+2] = [str(int(sec) * 3)]
        return array
    else:
      return array
```

```python
new_length = length
while new_length > 1:
  length = new_length
  for i in range(len(bracket)-1):
    try:
      score(i, bracket)
    except:
      pass
  new_length = len(bracket)
  if new_length == length:
    print(0)
    exit(0)
print(int(bracket[0]))
```
